```python
import cv2
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python


# Load your RGB renal image
image_path = './images_1024/VUHSK_2072_59.png'
image = cv2.imread(image_path)

# Convert the image to RGB (if not already)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Flatten the image to use as features for K-means
pixels = image_rgb.reshape((-1, 3))

# Specify the number of clusters (adjust as needed)
num_clusters = 5

# Apply K-means clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
labels = kmeans.fit_predict(pixels)

# Reshape the labels to the original image shape
segmentation_mask = labels.reshape(image_rgb.shape[:2])

# Display original image and generated segmentation mask
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(segmentation_mask, cmap='viridis')
plt.title('Generated Segmentation Mask')

plt.show()
```
